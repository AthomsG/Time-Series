```python
import re
from collections import defaultdict

def parse_input(input_string):
    pattern = r"City: (.*?)\n(.*?)(?=\nCity:|$)"
    city_matches = re.findall(pattern, input_string, re.DOTALL)

    data = defaultdict(list)
    for city_match in city_matches:
        city = city_match[0].strip()
        model_matches = re.findall(r"Model: (.*?)\nAIC: (.*?)\nBIC: (.*?)\n", city_match[1], re.DOTALL)
        for model_match in model_matches:
            model = model_match[0].strip()
            aic_str = model_match[1].strip()
            bic_str = model_match[2].strip()
            if aic_str and bic_str: # Check if strings are not empty
                aic = float(aic_str)
                bic = float(bic_str)
                score = aic + bic
                data[city].append((model, aic, bic, score))
    return data

def best_models(data):
    best_models = {}
    for city, models in data.items():
        best = sorted(models, key=lambda x: x[3])[:6]
        best_models[city] = best
    return best_models

def main():
    with open(path_to_file, 'r') as f:  # replace 'path_to_your_file.txt' with the path to your .txt file
        input_string = f.read()
    data = parse_input(input_string)
    best = best_models(data)
    for city, models in best.items():
        print(f"City: {city}\n")
        for model in models:
            print(f"{model_dict[model[0]]}, AIC: {model[1]}, BIC: {model[2]}")
        print()

main()
```
